```python
import pandas as pd

# Load the dataset
df = pd.read_csv("df_shortcuts.csv")

# Group the data by scenario and count the unique trajectory IDs in each group
trajectory_count_per_scenario = df.groupby('scenario')['trajectory_Id'].nunique()

# Print the count of trajectories per scenario
print("Trajectory count per scenario:")
print(trajectory_count_per_scenario)

```

```python
import matplotlib.pyplot as plt

# Define scenario labels and colors
scenario_labels = {
    1: 'light_noadapt',
    2: 'light_adapt',
    3: 'heavy_noadapt',
    4: 'heavy_adapt'
}

scenario_colors = {
    1: "#66B887",
    3: "#b084bf",
    2: "#008837",
    4: "#7b3294"
}

# Plot the count of trajectories per scenario
plt.figure(figsize=(10, 6))
bars = trajectory_count_per_scenario.plot(kind='bar', color=[scenario_colors[scenario] for scenario in trajectory_count_per_scenario.index])
plt.title('Trajectory with shortcuts Count per Scenario \n (Each scenario has a total of 216 trajectories)')
plt.xlabel('Scenario')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

# Update the x-axis tick labels
plt.xticks(range(len(scenario_labels)), [scenario_labels[scenario] for scenario in trajectory_count_per_scenario.index])

# Annotate each bar with its count
for bar in bars.patches:
    plt.annotate(str(bar.get_height()), xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()), xytext=(0, 3), 
                 textcoords="offset points", ha='center', va='bottom')

plt.show()
```
