```python
#add task id
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('df_phase_distance.csv')

# Plot navigating_distance distribution
plt.figure(figsize=(8, 6))
plt.hist(df['navigating_distance'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Navigating Distance')
plt.ylabel('Frequency')
plt.title('Distribution of Navigating Distance')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('df_phase_distance.csv')

# Plot navigating_distance distribution using scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df.index, df['navigating_distance'], color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Index')
plt.ylabel('Navigating Distance')
plt.title('Scatter Plot of Navigating Distance')
plt.grid(True)
plt.show()
```

```python
#density 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with the data
# Load your DataFrame
df = pd.read_csv('df_phase_distance.csv')

# Create a density plot using seaborn
plt.figure(figsize=(8, 6))
sns.kdeplot(df['navigating_distance'], shade=True)
plt.xlabel('Navigating Distance')
plt.ylabel('Density')
plt.title('Density Plot of Navigating Distance')
plt.show()
```
