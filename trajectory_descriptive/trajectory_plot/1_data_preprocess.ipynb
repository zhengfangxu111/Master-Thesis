```python
#________________add task id 
#clean_up phases
import pandas as pd
df_all = pd.read_csv('df_newphase.csv')
df_all['taskState'] = df_all['taskState'].fillna('')
# Function to assign taskId based on taskState for each participant and scenario
def assign_taskId(group):
    current_taskId = 0
    for index, row in group.iterrows():
        if row['taskState'] == 'taskAccepted':
            current_taskId += 1
        elif row['taskState'] == 'checkedIn':
            current_taskId = 0
        group.at[index, 'taskId'] = current_taskId
    return group

# Apply the function to each group (participant and scenario)
df_all = df_all.groupby(['participant', 'scenario']).apply(assign_taskId)

df_all.to_csv('df_ready.csv', index=False, encoding='utf-8')
```

```python
#add lockscreen mark
import pandas as pd


# Initialize the lockscreen column with zeros
df_lt['lockscreen'] = 0

# Initialize a variable to keep track of the lock screen state
lock_screen = False

# Iterate over the DataFrame rows
for index, row in df_lt.iterrows():
    # Check if mapInteractions column has 'mapLog:isScreenLocked+True'
    if row['mapInteractions'] == 'mapLog:isScreenLocked+True':
        # Set lock_screen to True
        lock_screen = True
    # Check if mapInteractions column has 'mapLog:isScreenLocked+False'
    elif row['mapInteractions'] == 'mapLog:isScreenLocked+False':
        # Set lock_screen to False
        lock_screen = False
    # Assign 1 to lockscreen column if lock_screen is True
    if lock_screen:
        df_lt.at[index, 'lockscreen'] = 1

# Print the updated DataFrame
df_lt.to_csv('df_lockmark.csv')
```
